# First event inspection

Load a GUNDAM engine and inspect the first event of the first model sample.

```python
nCpuThreads = 3
gundamLibPath = "/Users/nadrino/Documents/Work/Install/gundam/lib"
workDir = "/Users/nadrino/Documents/Work/Output/results/gundam/GundamInputOA2024"
configPath = "configOA2024.yaml"
overrideList = [
    "override/v12ProdRun45.yaml",
    "override/onlyFlux5.yaml",
    "override/noEigen.yaml",
]
dataType = "Toy"  # "Asimov", "Toy", or "RealData"
seed = 12345
```

```python
import sys
from pathlib import Path

import numpy as np

repoRoot = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
srcPath = repoRoot / "src"
if srcPath.exists() and str(srcPath) not in sys.path:
    sys.path.insert(0, str(srcPath))

from gundam_interface import GundamInterface, GundamLoader, GundamRuntime
```

```python
np.random.seed(seed)

runtime = GundamRuntime(
    loader=GundamLoader(gundamLibPath=gundamLibPath),
    workDir=workDir,
    nCpuThreads=nCpuThreads,
    configPath=configPath,
    overrideList=overrideList,
    dataType=dataType,
    randomSeed=seed,
)

runtime.toDict(includeConfigJsonString=False)
```

```python
gundam = GundamInterface(runtime)
gundam.configure()
gundam.initialize()
```

```python
sampleIndex = 0
eventIndex = 0

sample = gundam.modelSamples[sampleIndex].handle
events = sample.getEventList()
event = events[eventIndex]

indices = event.getIndices()

print(f"Sample index: {sampleIndex}")
print(f"Sample name:  {sample.getName()}")
print(f"Event index:  {eventIndex}")
print(f"Event weight: {event.getEventWeight():.8g}")
print(f"Entry:        {indices.entry}")
print(f"Bin:          {indices.bin}")

print(event.getSummary())
```

```python
variables = event.getVariables()
varList = variables.getVarList()

print(f"Number of variables: {len(varList)}")
for index, variable in enumerate(varList[:10]):
    print(f"{index:3d}: {variable.getVarAsDouble():.8g}")
```

```python
names = variables.getNameList()
for index, name in enumerate(names):
    print(f"{index:3d}: {name}, {variables.fetchVariable(name).getVarAsDouble():.8g}")
```
